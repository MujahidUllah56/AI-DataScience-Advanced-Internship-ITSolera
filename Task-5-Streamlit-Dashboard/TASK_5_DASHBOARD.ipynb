```python
%%writefile app.py
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt

# ================= PAGE CONFIG =================
st.set_page_config(
    page_title="Executive Sales Dashboard",
    page_icon="ğŸ“Š",
    layout="wide"
)

# ================= TITLE =================
st.markdown(
    """
    <h1 style='text-align:center;'>ğŸ“Š Executive Sales Dashboard</h1>
    <h4 style='text-align:center; color:gray;'>
    Global Superstore â€“ Business Performance Analysis
    </h4>
    """,
    unsafe_allow_html=True
)

st.markdown("---")

# ================= LOAD & CLEAN DATA =================
@st.cache_data
def load_and_clean_data():
    df = pd.read_csv("Global_Superstore.csv", encoding="latin1")

    # --- Select relevant columns ---
    df = df[
        [
            "Order Date", "Region", "Category", "Sub-Category",
            "Customer Name", "Sales", "Profit"
        ]
    ]

    # --- Convert dates ---
    df["Order Date"] = pd.to_datetime(df["Order Date"], errors="coerce")

    # --- Convert numeric columns ---
    df["Sales"] = pd.to_numeric(df["Sales"], errors="coerce")
    df["Profit"] = pd.to_numeric(df["Profit"], errors="coerce")

    # --- Remove missing values ---
    df.dropna(subset=["Sales", "Profit", "Order Date"], inplace=True)

    # --- Remove duplicate rows ---
    df.drop_duplicates(inplace=True)

    return df

df = load_and_clean_data()

# ================= SIDEBAR FILTERS =================
st.sidebar.title("ğŸ” Dashboard Filters")

region = st.sidebar.multiselect(
    "Region",
    options=sorted(df["Region"].unique()),
    default=sorted(df["Region"].unique())
)

category = st.sidebar.multiselect(
    "Category",
    options=sorted(df["Category"].unique()),
    default=sorted(df["Category"].unique())
)

sub_category = st.sidebar.multiselect(
    "Sub-Category",
    options=sorted(df["Sub-Category"].unique()),
    default=sorted(df["Sub-Category"].unique())
)

# ================= APPLY FILTERS =================
filtered_df = df[
    (df["Region"].isin(region)) &
    (df["Category"].isin(category)) &
    (df["Sub-Category"].isin(sub_category))
]

# ================= KPI CALCULATIONS =================
total_sales = filtered_df["Sales"].sum()
total_profit = filtered_df["Profit"].sum()
total_customers = filtered_df["Customer Name"].nunique()

# ================= KPI DISPLAY =================
k1, k2, k3 = st.columns(3)

k1.metric("ğŸ’° Total Sales", f"${total_sales:,.0f}")
k2.metric("ğŸ“ˆ Total Profit", f"${total_profit:,.0f}")
k3.metric("ğŸ‘¥ Total Customers", total_customers)

st.markdown("---")

# ================= CHARTS =================
c1, c2 = st.columns(2)

with c1:
    st.subheader("Sales by Region")
    if filtered_df.empty:
        st.warning("No data available for selected filters.")
    else:
        region_sales = filtered_df.groupby("Region")["Sales"].sum().sort_values()
        fig, ax = plt.subplots()
        region_sales.plot(kind="barh", ax=ax)
        ax.set_xlabel("Sales")
        ax.set_ylabel("Region")
        st.pyplot(fig)

with c2:
    st.subheader("Top 5 Customers by Sales")
    if filtered_df.empty:
        st.warning("No data available for selected filters.")
    else:
        top_customers = (
            filtered_df.groupby("Customer Name")["Sales"]
            .sum()
            .sort_values(ascending=True)
            .tail(5)
        )
        fig2, ax2 = plt.subplots()
        top_customers.plot(kind="barh", ax=ax2)
        ax2.set_xlabel("Sales")
        ax2.set_ylabel("Customer")
        st.pyplot(fig2)

st.markdown("---")

# ================= FILTER SUMMARY =================
st.subheader("ğŸ“Œ Filter Summary")
st.write(f"Total Rows After Filtering: {filtered_df.shape[0]}")

# ================= CLEAN DATA PREVIEW =================
st.subheader("ğŸ“„ Cleaned & Filtered Dataset")

if filtered_df.empty:
    st.warning("âš ï¸ No data available for selected filters.")
else:
    st.dataframe(filtered_df, use_container_width=True)
```
